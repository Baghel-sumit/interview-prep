> The name of a class starts with a capital letter

Inheritance: With inheritance one class can derive the properties of another class

Types of inheritance

1. Single Inheritance: the child inherits from only 1 parent class
2. Multiple Inheritance: the child inherits from more than 1 parent class
3. Multi-level Inheritance: we have parent, child, grand-child relationship
4. Hybrid Inheritance: 

```python
# class declaration

class Phone:

  def make_call(self):
    print("making phone call")
  def play_game(self):
    print("starting the game")

phone1 = Phone() # initiating an object of the class

phone1.make_call()
phone1.play_game()
```

```python
# class declaration with params

class Phone:

  def set_color(self, color):
    self.color = color

  def set_cost(self, cost):
    self.cost = cost
  
  def show_color(self):
    return self.color

  def show_cost(self):
    return self.cost

  def make_call(self):
    print("Making a call")
  
  def play_game(self):
    print("playing game")
  
  def show_self_type(mine):
    print("type", type(mine))

phone1 = Phone()

phone1.set_color("red")
phone1.set_cost(23049734)

print(phone1.show_color())
print(phone1.show_cost())
phone1.make_call()
phone1.play_game()

phone1.show_self_type()
```

```python
# constructor declaration

class Employee:
  def __init__(self, name: str, age: int, salary: int, gender: str = "male") -> None:
    self.name = name
    self.age = age
    self.salary = salary
    self.gender = gender

  def get_information(self):
    dict = {
      "name": self.name,
      "age": self.age,
      "salary": self.salary,
      "gender": self.gender
    }
    print(dict)

employee1 = Employee("sumit", 23, 4334, "M")

employee1.get_information()
```

```python
# inheritance, overriding constructor

class Vehicle:

  def __init__(self, mileage, cost):
    self.mileage = mileage
    self.cost = cost
  
  def show_details(self):
    dict = {
      "mileage": self.mileage,
      "cost": self.cost
    }
    return dict

class Car(Vehicle):

  def __init__(self, mileage, cost, tyres, fuelType):
    super().__init__(mileage, cost)
    self.tyres = tyres
    self.fuelType = fuelType

  def show_car(self):
    dict = [self.mileage, self.cost, self.tyres, self.fuelType]
    print(dict)

car1 = Car(200, 400000, "mrf", "petrol")

car1.show_car()
```

```python
# Multiple inheritance

class Parent1:
  
  def assign_string_one(self, str1):
    self.str1 = str1
  
  def show_string_one(self):
    return self.str1

class Parent2:

  def assign_string_two(self, str2):
    self.str2 = str2
  
  def show_string_two(self):
    return self.str2

class Derived(Parent1, Parent2):

  def assign_string_three(self, str3):
    self.str3 = str3
  
  def show_string_three(self):
    return self.str3
  
d1 = Derived()

d1.assign_string_one("one")
d1.assign_string_two("two")
d1.assign_string_three("three")

print(d1.show_string_one())
print(d1.show_string_two())
print(d1.show_string_three())
```

```python
# Multi- level inheritance

class Parent:

  def assign_name(self, name):
    self.name = name
  
  def show_name(self):
    return self.name

class Child(Parent):

  def assign_age(self, age):
    self.age = age

  def show_age(self):
    return self.age

class Grand_child(Child):

  def assign_gender(self, gender):
    self.gender = gender
  
  def show_gender(self):
    return self.gender
  
g1 = Grand_child()

g1.assign_name("sumit")
g1.assign_age(23)
g1.assign_gender("male")

print(g1.show_age())
print(g1.show_name())
print(g1.show_gender())
```
